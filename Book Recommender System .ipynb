# Popularity Based Recommender System
Top 50 book,highest average rating,min. of 250 ratings

```python
ratings_with_name = books.merge(ratings,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.sample(5)
```

```python
num_rat= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rat.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rat
```

```python
avg_rat= ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rat.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rat
```

```python
popular_df = num_rat.merge(avg_rat,on='Book-Title')
popular_df
```

```python
popular_df= popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.head()
```

```python
import pickle as pk
pk.dump(popular_df,open("popularbooks.pkl",'wb'))
```
