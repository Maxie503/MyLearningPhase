```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df =pd.read_csv(r"C:\Users\U6056186\Downloads\student_clustering.csv")
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['iq'])
plt.show()
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss[1:])
```

```python
X=df.iloc[:,:].values
km= KMeans(n_clusters=4)
ymeans = km.fit_predict(X)
```
