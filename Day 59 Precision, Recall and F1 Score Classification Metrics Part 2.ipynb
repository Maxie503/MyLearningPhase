```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression, make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=500, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_,lr.intercept_
y_pred1 = lr.predict(X_test)
```

```python
m = -(lr.coef_[0]/lr.coef_[1])
b = -(lr.intercept_/lr.coef_[1])
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
plt.figure(figsize=(10,6))
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
X1, y1 = X, y = make_classification(n_samples=500, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=2)
plt.figure(figsize=(10,6))
plt.scatter(X1[:,0],X1[:,1],c=y1,cmap='winter',s=100)
```

```python
X_train,X_test,y_train,y_test2 = train_test_split(X1,y1,test_size=0.2,random_state=4)
```

```python
lr2=LinearRegression()
```

```python
lr2.fit(X_train,y_train)
```

```python
m = -(lr2.coef_[0]/lr2.coef_[1])
b = -(lr2.intercept_/lr2.coef_[1])
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
plt.figure(figsize=(10,6))
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X1[:,0],X1[:,1],c=y1,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
y_pred2 = lr2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Regression1",accuracy_score(y_test,y_pred1))
print("Regression1",accuracy_score(y_test2,y_pred2))
```
