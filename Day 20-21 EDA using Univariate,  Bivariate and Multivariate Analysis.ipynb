# Day 21 :  Bivariate and Multivariate Analysis\


```python
titanic=pd.read_csv(r'C:\Users\U6056186\Downloads\train.csv')
```

```python
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
flight.head()
```

```python
iris.head()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,hue=tips.sex,
               size=tips.size,style=tips.smoker)
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,size=tips.size)
```

```python
titanic.head()
```

```python
sns.barplot(x=titanic.Pclass,y=titanic.Age,hue=titanic.Sex)
```

```python
sns.barplot(x=titanic.Pclass,y=titanic.Fare,hue=titanic.Sex)
```

```python
sns.boxplot(x=titanic.Sex,y=titanic.Age,hue=titanic.Survived)
```

```python
sns.displot(x=titanic[(titanic.Survived)==0]['Age'],kind='kde')
sns.displot(x=titanic[(titanic.Survived)==1]['Age'],kind='kde')
plt.show()
```

```python
pd.crosstab(titanic['Pclass'],titanic.Survived)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic.Survived))
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic.Survived))
```

```python
titanic.groupby('Pclass').mean()['Survived'].plot(kind='bar')
```

```python
titanic.groupby('Sex').mean()['Survived'].plot(kind='bar')
```

```python
titanic.groupby('Embarked').mean()['Survived'].plot(kind='bar')
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic.Survived))
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic.Survived))
```

```python
sns.pairplot(iris,hue='species')
```
